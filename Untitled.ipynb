```python
num_feat = ['Age','Vintage']
cat_feat = ['Gender', 'Driving_License', 'Previously_Insured', 'Vehicle_Age_lt_1_Years', 
            'Vehicle_Age_gt_2_Years', 'Vehicle_Damage_Yes', 'Region_Code', 'Policy_Sales_Channel']
```

```python
df['Gender'] = df['Gender'].map({'Female':0, 'Male': 1}).astype(int)
df.head(2)
```

```python
df.info()
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
df = df.rename(columns = {'Vehicle_Age_< 1 Year':'Vehicle_Age_lt_1_Years', 'Vehicle_Age_> 2 Years':'Vehicle_Age_gt_2_Years'})
```

```python
df['Vehicle_Age_lt_1_Years'] = df['Vehicle_Age_lt_1_Years'].astype('int')
df['Vehicle_Age_gt_2_Years'] = df['Vehicle_Age_gt_2_Years'].astype('int')
df['Vehicle_Damage_Yes'] = df['Vehicle_Damage_Yes'].astype('int')
```

```python
for col in cat_feat:
    df[col] = df[col].astype('str')
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler, RobustScaler
```

```python
ss = StandardScaler()
df[num_feat] = ss.fit_transform(df[num_feat])

mm = MinMaxScaler()
df[['Annual_Premium']] = mm.fit_transform(df[['Annual_Premium']])

id = df.id
df = df.drop('id', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_target = df['Response']
train = df.drop(['Response'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(train, train_target, test_size=0.2, random_state=0)
```

```python
train_target
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=500, criterion='gini')
model = rf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_search = {'criterion': ['entropy', 'gini'],
                'max_depth': [2,3,4,5,6,7,10],
                'min_samples_leaf': [4,6,8],
                'n_estimators':[300]}
model = RandomizedSearchCV(estimator=rf, param_distributions=random_search, n_iter=10,
                          cv=4, verbose=1, random_state=101, n_jobs=-1)
model.fit(X_train, y_train)
```

```python
print("shape:", y_train.shape)
print("Dtype:", y_train.dtype)
print("Unique:", pd.Series(y_train).unique())
print("Types of Unique Values:", [type(x) for x in pd.Series(y_train).unique()])
```

```python
best_params = model.best_params_
print("Best Hyperparameters")
print(best_params)
```

```python
best_rf = RandomForestClassifier(n_estimators=300, min_samples_leaf=8, max_depth=10, criterion='entropy')
best_rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = best_rf.predict(X_test)
print(classification_report(y_test, y_pred))
```
